<h1/>Задачи. Модуль "Разделяй и властвуй"</h1>

```python
import sys
from functools import partial


def binSearch(A, key):
    n = len(A)
    L, R = 0, n-1
    while L <= R:
        m = L + (R - L) // 2
        if A[m] == key:
            return m+1
        elif A[m] > key:
            R = m - 1
        else:
            L = m + 1
    return -1


def main():
    reader = (map(int, line.split()) for line in sys.stdin)
    n, *A = next(reader)
    k, *keys = next(reader)
    assert n == len(A)
    assert k == len(keys)

    binSearchA = partial(binSearch, A)
    print(*list(map(binSearchA, keys)), sep=' ')


def test():
    n, A = 5, [1, 5, 8, 12, 13]
    k, keys = 5, [8, 1, 23, 1, 11]
    result = [3, 1, -1, 1, -1]
    binSearchA = partial(binSearch, A)
    assert list(map(binSearchA, keys)) == result
    assert binSearch([], 1) == -1
    assert binSearch([1], 1) == 1
    assert binSearch([1, 2], 5) == -1


if __name__ == '__main__':
    main()
```

```python
# РЕШЕНИЕ №1

import sys

inv = 0


def Merge(L, R):
    global inv
    i = j = 0
    res = []
    while i < len(L) and j < len(R):
        if L[i] <= R[j]:
            res.append(L[i])
            i += 1
        else:
            res.append(R[j])
            j += 1
            inv += len(L[i:])
    res += L[i:] + R[j:]
    return res


def MergeSort(A):
    if len(A) <= 1:
        return A
    else:
        L = A[:len(A) // 2]
        R = A[len(A) // 2:]
    return Merge(MergeSort(L), MergeSort(R))


def main():
    reader = (map(int, line.split()) for line in sys.stdin)
    [n], A = next(reader), list(next(reader))
    MergeSort(A)
    print(inv)


def test():
    n, A = 5, [2, 3, 9, 2, 9]
    assert MergeSort(A) == [2, 2, 3, 9, 9], 2


if __name__ == '__main__':
    main()
```

```python
# РЕШЕНИЕ №2

inv = 0

def merge(left, right):
    global inv
    lst = []
    while left and right:
        if left[0] <= right[0]:
            lst.append(left.pop(0))
        else:
            lst.append(right.pop(0))
            inv += len(left)
    if left:
        lst.extend(left)
    if right:
        lst.extend(right)
    return lst

def mergesort(lst):

    length = len(lst)
    if length >= 2:
        mid = int(length / 2)
        lst = merge(mergesort(lst[:mid]), mergesort(lst[mid:]))
    return lst

n = int(input())
A = list(map(int, input().split()))
mergesort(A)
print(inv)
```

```python
# http://www.cyberforum.ru/cpp-beginners/thread1163749.html

import sys
from random import choice


def QuickSort(A):
    if len(A) <= 1:
        return A
    else:
        q = choice(A)
        L = []
        M = []
        R = []
        for elem in A:
            if elem < q:
                L.append(elem)
            elif elem > q:
                R.append(elem)
            else:
                M.append(elem)
        return QuickSort(L) + M + QuickSort(R)


def main():
    reader = (map(int, line.split()) for line in sys.stdin)
    n, m = next(reader)
    L = []
    R = []
    for i in range(n):
        begin, end = next(reader)
        L.append(begin)
        R.append(end)
    L = QuickSort(L)
    R = QuickSort(R)
    for point in next(reader):
        acc = 0
        left = 0
        right = len(L) - 1
        while left <= right:
            m = left + (right - left) // 2
            if L[m] > point:
                right = m - 1
            else:
                left = m + 1

        acc += right + 1
        left = 0
        right = len(R) - 1
        while left <= right:
            m = left + (right - left) // 2
            if R[m] < point:
                left = m + 1
            else:
                right = m -1
        acc -= right + 1
        print(acc, end=' ')


if __name__ == '__main__':
    main()
```

```python
import sys


def CountSort(A):
    '''Устойчивая сортировка подсчетом

    :param A: массив А должен содержать целые числа от 1 до 10
    :return: возвращает упорядоченную по неубыванию последовательность этих чисел

    '''

    B = [0] * 10
    for j in A:
        B[j-1] += 1
    for i in range(1, 10):
        B[i] += B[i-1]
    result = [None] * len(A)
    for k in reversed(A):
        result[B[k-1]-1] = k
        B[k-1] -= 1
    return result


def main():
    reader = (map(int, line.split()) for line in sys.stdin)
    [n], A = next(reader), list(next(reader))
    print(*CountSort(A))


if __name__ == '__main__':
    main()
```

```python
import sys
from bisect import bisect_left

def with_bisect(xs, query):
    lo = bisect_left(xs, query)
    # i < lo : xs[i] < query
    # i > lo : xs[i] >= query
    if lo < len(xs) and xs[lo] == query:
        return lo + 1  # 1 - based
    else:
        return -1


def bin_search(xs, query):
    # Invariant: lo <= pos <= hi
    lo, hi = 0, len(xs) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if query < xs[mid]:
            hi = mid - 1      # [lo, mid - 1]
        elif query > xs[mid]:
            lo = mid + 1   # [mid + 1, hi)
        else:
            return mid + 1 # 1-based

    return -1


def find_pos(xs, query):
    try:
        return xs.index(query) + 1
    except ValueError:
        return -1

def test():
    assert bin_search([], 42) == -1
    assert bin_search([42], 42) == 1
    assert bin_search([42], 24) == -1
    assert [bin_search([1, 5, 8, 12, 13], i) \
            for i in [8, 1, 23, 1, 11]] == [3, 1, -1, 1, -1]


def main():
    reader = (map(int, line.split()) for line in sys.stdin)
    n, *xs = next(reader)
    k, *queries = next(reader)
    for query in queries:
        print(bin_search(xs, query), end=" ")


if __name__ == '__main__':
    test()
```
