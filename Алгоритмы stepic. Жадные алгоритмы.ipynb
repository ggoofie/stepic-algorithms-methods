```python
from operator import itemgetter

n = int(input().rstrip())
sec = sorted([list(map(int, input().split())) for i in range(n)], key=itemgetter(1))
i = 0
l, r = 0, 1
result = []
while i < n:
        result.append(sec[i][r])
        i += 1
        while i < n and result[-1] >= sec[i][l]:
            i += 1
print(len(result), *result, sep='\n')
```

```python
n, W = map(int, input().split())
cw = sorted([list(map(int, input().split())) for i in range(n)], \
            key=lambda thing: thing[0]/thing[1], reverse=True)
c, w = 0, 1
result = 0
for thing in cw:
    if W >= thing[w]:
        W -= thing[w]
        result += thing[c]
    else:
        result += (thing[c]/thing[w]) * W
        break
print('%.3f' % result)    
```

```python
from heapq import heappush, heappop, heapify
from collections import Counter, namedtuple


class Node(namedtuple('Node', ['left', 'right'])):
    def walk(self, code, acc):
        self.left.walk(code, acc + '0')
        self.right.walk(code, acc + '1')

        
class Leaf(namedtuple('Leaf', ['char'])):    
    def walk(self, code, acc):
        if acc:
            code[self.char] = acc
        else:
            code[self.char] = '0'

            
def Huffman(s):
    h = []
    for char, freq in Counter(s).items():
        h.append((freq, len(h), Leaf(char)))
    heapify(h)
    while len(h) > 1:
        i = heappop(h)
        j = heappop(h)
        heappush(h, (i[0]+j[0], len(h), Node(i[2], j[2])))
    code = {}
    if h:
        tree = h[0][2]
        tree.walk(code, '')
    return code

s = input().rstrip()
code = Huffman(s)
encoded = "".join(code[char] for char in s)
print(len(code), len(encoded))
for i in sorted(code):
    print('%s: %s' % (i, code[i]))
print(encoded)
```

```python
# https://www.youtube.com/watch?v=qkLLcdgJj_o&feature=youtu.be&t=3327

class heap:
    def __init__(self):
        self.a = [None]
        
    def siftup(self, i):
        while i > 1 and self.a[i//2] < self.a[i]:
            self.a[i], self.a[i//2] = self.a[i//2], self.a[i]
            i = i//2

    def insert(self, v):
        self.a.append(v)
        self.siftup(len(self.a)-1)

    def siftdown(self, i):
        while 2*i <= len(self.a)-1:
            j = i
            if self.a[2*i] > self.a[j]:
                j = 2*i
            if 2*i+1 <= len(self.a)-1 and self.a[2*i+1] > self.a[j]:
                j = 2*i+1
            if j == i:
                break
            self.a[i], self.a[j] = self.a[j], self.a[i]
            i = j
        
    def extractMax(self):
        ex_max = self.a[1]
        self.a[1] = self.a[len(self.a)-1]
        del self.a[len(self.a)-1]
        self.siftdown(1)
        return ex_max
        
            
h = heap()
commands = [input().split() for i in range(int(input()))]
for k in commands:
    if len(k) == 2:
        h.insert(int(k[1]))
    else:
        print(h.extractMax())
```

<h2>Задача о рюкзаке</h2>

```python
import sys
import heapq


def fractional_knapsack(capacity, values_and_weights):
    order = [(-v / w, w) for v, w in values_and_weights] # "-" перед v инвертиртирует поведение кучи,
    heapq.heapify(order)                                 # в heapq мин-куча, сдеалали макс-кучу

    acc = 0
    while order:
        v_per_w, w = heapq.heappop(order)
        can_take = min(w, capacity)
        acc += -v_per_w * can_take
        capacity -= can_take
        
    return acc


def main():
    reader = (tuple(map(int, line.split())) for line in sys.stdin)
    n, capacity = next(reader)
    values_and_weights = list(reader)
    assert len(values_and_weights) == n
    opt_value = fractional_knapsack(capacity, values_and_weights)
    print('{:.3f}'.format(opt_value))
   

def test():
    # краевые случая и пример со степика
    assert fractional_knapsack(0, [(60, 20)]) == 0.0
    assert fractional_knapsack(25, [(60, 20)]) == 60.0
    assert fractional_knapsack(25, [(60, 20), (0, 100)]) == 60.0
    assert fractional_knapsack(25, [(60, 20), (50, 50)]) == 60.0 + 5.0
    assert fractional_knapsack(50, [(60, 20), (100, 50), (120, 30)]) == 180.0
    
    # замерим время работы вычислительной части
    from random import randint
    from timing import timed            # timed писали в модуле "Введение"
    
    for attempt in range(100):
        n = randint(1, 1000)
        capacity = randint(0, 2 * 10**6)
        values_and_weights = []
        for i in range(n):
            values_and_weights.append(
                (randint(0, 2 * 10**6), randint(1, 2 * 10**6)))
            
        t = timed(fractional_knapsack, capacity, values_and_weights)
        assert t < 5
 

if __name__ == '__main__':
    test()
```

<h2>Коды Хаффмана</h2>

```python
import heapq
from collections import Counter, namedtuple


class Node(namedtuple("Node", ["left", "right"])):
    def walk(self, code, acc):
        self.left.walk(code, acc + '0')
        self.right.walk(code, acc + '1')
 

class Leaf(namedtuple("Leaf", ["char"])):
    def walk(self, code, acc):
        code[self.char] = acc or "0"


def huffman_encode(s):
    h = []
    for ch, freq in Counter(s).items():
        h.append((freq, len(h), Leaf(ch)))
        
    heapq.heapify(h)
    
    count = len(h)
    while len(h) > 1:
        freq1, _count1, left = heapq.heappop(h)  # символ _ - слабый индикатор того, что имя используется для
        freq2, _count2, right = heapq.heappop(h)  # внутренних нужд. Например, from M import * не будет импортировать объекты,
        heapq.heappush(h, (freq1 + freq2, count, Node(left, right)))  # чьи имена начинаются с символа подчеркивания.
        count += 1
        
    code = {}
    if h:
        [(_freq, _count, root)] = h
        root.walk(code, "")
    return code


def huffman_decode(code, encoded):
    i, j = 0, 1
    result = ''
    while j <= len(encoded):
        select = list(filter(lambda key: code[key].startswith(encoded[i:j]), code))
        if len(select) == 1:
            result += select[0]
            i, j = j, j+1
        else:
            j += 1
    return result


def main():
    s = input()
    code = huffman_encode(s)
    encoded = "".join(code[ch] for ch in s)
    print(len(code), len(encoded))
    for ch in sorted(code):
        print("{}: {}".format(ch, code[ch]))
    print(encoded)
    

def test(n_iter=100):  # тестировать будем исходя из того, что кодирование и декодирование взаимнообратны
    import random
    import string
    
    for i in range(n_iter):
        length = random.randint(0, 32)
        s = "".join(random.choice(string.ascii_letters) for _ in range(length))
        code = huffman_encode(s)
        encoded = "".join(code[ch] for ch in s)
        assert huffman_decode(code, encoded) == s

        
if __name__ == '__main__':
    test()
```
